```python
# Import Library
import requests
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
from datetime import datetime
```

```python
# Import excel file containing url list (retrieved from sitemap of the main url)
df = pd.read_excel(r"~/Documents/Datasets/kakao/kakao_webtoon.xlsx")
df.head()
```

```python
# Try 1 iteration to understand the page source code structure

url1 = df["Content"][1]
print(url1)
r1 = requests.get(url1)
soup1 = BeautifulSoup(r1.content, 'html5lib')
print(soup1.prettify())
```

```python
#loop thru to get title, author_group, genre, views, and likes
results = soup1.find_all("p")
result_text = []
for i in range(len(results)):
    the_text = results[i].text
    if the_text != "" and the_text != "0":
        result_text.append(results[i].text)
        
result_text
```

```python
desc = soup1.find("meta", attrs = {'name':'description'})["content"]
print(desc) #description
key = soup1.find("meta", attrs = {'name':'keywords'})["content"]
print(key) #writer, artist, original work (for adaptation work), publisher (studio) 
```

```python
imgurl = soup1.find("meta", attrs = {"property":"og:image"})["content"]
print(imgurl) #imgurl
```

```python
by_titles = []
n = len(df["Content"])
```

```python
n
```

```python
# Iterate for all series in the list
for i in range(n):
    the_url = df.loc[i,"Content"]
    result = requests.get(the_url)
    soup = BeautifulSoup(result.content, 'html5lib')
    #create dictionary item for this loop
    by_title = {'Id':df.loc[i,"Id"]}
    #get title, author_group, genre, views, and likes
    subresults = soup.find_all("p")
    subresult_text = []
    for j in range(len(subresults)):
        the_text = subresults[j].text
        if the_text != "" and the_text != "0":
            subresult_text.append(the_text)
    by_title['subresult_text'] = subresult_text[:]
    by_title['title'] = subresult_text[0]
    by_title['author_group'] = subresult_text[1]
    by_title['genre'] = subresult_text[2]
    by_title['views'] = subresult_text[3]
    by_title['likes'] = subresult_text[4]
    #get description
    by_title['desc'] = soup.find("meta", attrs = {'name':'description'})["content"]
    #get keywords
    by_title['key'] = soup.find("meta", attrs = {'name':'keywords'})["content"]
    #append dictionary from this iteration to main list "by_titles"
    by_titles.append(by_title)
    #print iteration number to monitor the iteration process
    print(i)
```

```python
# Iterate for all series in the list
for i in range(571,n):
    the_url = df.loc[i,"Content"]
    result = requests.get(the_url)
    soup = BeautifulSoup(result.content, 'html5lib')
    #create dictionary item for this loop
    by_title = {'Id':df.loc[i,"Id"]}
    #get imgurl
    by_title['imgurl'] = soup.find("meta", attrs = {"property":"og:image"})["content"]
    #append dictionary from this iteration to main list "by_titles"
    by_titles.append(by_title)
    #print iteration number to monitor the iteration process
    print(i)
```

```python
# Transform dict to dataframe

df_detail2 = pd.DataFrame.from_dict(by_titles)
print(df_detail2.shape)
df_detail2.tail()
```

```python
def string_to_num(text):
    if "K" in text:
        return int(float(text.split("K")[0])*1000)
    elif "M" in text:
        return int(float(text.split("M")[0])*1000000)
    else:
        return int(text)
```

```python
df_main["views"] = df_main["views"].str.replace(",",".").apply(string_to_num)
df_main["likes"] = df_main["likes"].str.replace(",",".").apply(string_to_num)
df_main.tail()
```
